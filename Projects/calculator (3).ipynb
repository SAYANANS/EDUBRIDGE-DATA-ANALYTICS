```python
from tkinter import *
```

```python
window=Tk()
window.title("calculator")
window.configure(bg="grey")
T=Entry(window,width=35,bd=5,font=("Arial,25"))
T.grid(row=0,column=0,columnspan=4,padx=10,pady=10)
def Button_Clicker(number):
    new_number=T.get()+str(number)
    T.delete(0,END)
    T.insert(0,new_number)
    
def Button_Clear():
    T.delete(0,END)
    
def Button_Add(first_num): 
    global f_num
    global math
    math="Addition"
    f_num=first_num
    T.delete(0,END)
    
def Button_Mul(first_num):
    global f_num
    global math
    math="Multiplication"
    f_num=first_num
    T.delete(0,END)
    
def Button_Sub(first_num):
    global f_num
    global math
    math="Subtraction"
    f_num=first_num
    T.delete(0,END)
    
def Button_Div(first_num):
    global f_num
    global math
    math="Division"
    f_num=first_num
    T.delete(0,END)
    
def Button_Perc(first_num):
    global f_num
    global math
    math="Percentage"
    f_num=first_num
    T.delete(0,END) 
    
def Button_Equal(second_num):
    num_1 = f_num
    if math == "Addition":
        T.delete(0,END)
        T.insert(0,int(num_1)+int(second_num))
    elif math == "Multiplication":
        T.delete(0,END)
        T.insert(0,int(num_1)*int(second_num)) 
    elif math == "Subtraction":
        T.delete(0,END)
        T.insert(0,int(num_1)-int(second_num))
    elif math == "Division":
        T.delete(0,END)
        T.insert(0,int(num_1)/int(second_num))
    elif math == "Percentage":
        T.delete(0,END)
        T.insert(0,int(num_1)/100)   
        
    
        
    
    
    
    
button_1 = Button(window, text="1",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Clicker(1))
button_2 = Button(window, text="2",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Clicker(2))
button_3 = Button(window, text="3",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Clicker(3))
button_4 = Button(window, text="4",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Clicker(4))
button_5 = Button(window, text="5",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Clicker(5)) 
button_6 = Button(window, text="6",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Clicker(6))
button_7 = Button(window, text="7",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Clicker(7))
button_8 = Button(window, text="8",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Clicker(8))
button_9 = Button(window, text="9",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Clicker(9))
button_0 = Button(window, text="0",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Clicker(0))
button_0.grid(row=4, column=1)
button_1.grid(row=3, column=0)
button_2.grid(row=3, column=1)
button_3.grid(row=3, column=2)

button_4.grid(row=2, column=0)
button_5.grid(row=2, column=1)
button_6.grid(row=2, column=2)

button_7.grid(row=1, column=0)
button_8.grid(row=1, column=1)
button_9.grid(row=1, column=2)

button_clear=Button(window,text="clear",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Clear())
button_clear.grid(row=4, column=0)
button_add=Button(window,text="+",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Add(T.get()))
button_add.grid(row=4, column=2)
button_mul=Button(window,text="*",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Mul(T.get()))
#button_mul.grid(row=4,column=2)
button_sub=Button(window,text="-",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Sub(T.get()))
button_sub.grid(row=1, column=3)
button_div=Button(window,text="/",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Div(T.get()))
button_div.grid(row=2, column=3)
button_equal=Button(window,text="=",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Equal(T.get()))
button_equal.grid(row=4, column=3)
button_perc=Button(window,text="%",bg="black",fg="white",padx=70, pady=20,bd=3,font=("calibri",10), command=lambda: Button_Perc(T.get()))
button_perc.grid(row=3, column=3)

window.mainloop()
```
