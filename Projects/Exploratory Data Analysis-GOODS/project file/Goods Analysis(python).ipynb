##  Fruit & Veg prices(2017-2022)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
g=pd.read_csv('fruitvegprices-2017_2022.csv')
```

### Query to find the sum of duplicated values

```python
g.duplicated().sum()
```

### Query to sort data according to price

```python
g.sort_values(by=['Price'],inplace=True)
```

```python
g
```

### Query to rename column names

```python
g.rename(columns={'category':'Category','item':'Item','variety':'Variety','date':'Date','price':'Price','unit':'Unit'},inplace=True)
```

```python
g
```

### Query to create a pivot table for max price of each category

```python
g0=g.copy()
g0=g0.pivot_table('Price',columns='Category',aggfunc='max')
g0
```

### Query to create a pivot table for max price of each item 

```python
g1=g.copy()
g1=g1.pivot_table('Price',columns=['Item'],aggfunc='max')
g1
```

### Query to group Item with their respective Variety,Unit and Price

```python
g2=g.copy()
g2=g2.groupby(['Item','Variety','Unit'])['Price'].min()
g2=pd.DataFrame(g2)
g2
```

### Plot to know items and its counts

```python
plt.figure(figsize=(14,8))
sns.countplot(x=g['Item'])
plt.xlabel('ITEMS',fontsize=15,color="Blue")
plt.ylabel('QUANTITY',fontsize=15,color="Blue")
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

### Pie chart to find the count of each category

```python
plt.title('Category',fontsize=20,color='Blue')
g['Category'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True)
plt.show()
```

### Query to take 50 samples from data

```python
g3=g.sample(50)
g3
```

### Graph to show category wise price for each item

```python
plt.figure(figsize=(16,10))
plt.title('Category wise Price for each Item')
sns.barplot(x='Item',y='Price',data=g3,hue='Category',palette='rocket_r')
plt.xticks(rotation=90)
plt.xlabel('Region',fontsize=15)
plt.show()
```

### Query to sort data according to the date

```python
g4=g3.sort_values(by='Date')
```

```python
g4
```

### Query to find Details of apple

```python
g5=g4[(g4['Item']=='apples')]
```

```python
g5
```

### Line graph to show the Variation in the price of apple according to date

```python
plt.figure(figsize=(15,4))
sns.lineplot(x='Date',y='Price',data=g5,color='r',label='Price')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
