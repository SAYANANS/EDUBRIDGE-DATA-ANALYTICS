## Query to create a pivot table for max price of each item

```r
d %>% group_by(item) %>% summarize(max_price = max(price))
```

## Query to create a pivot table for min price of each item

```r
g1=d %>% group_by(item,variety,unit) %>% summarise(min_price = min(price))
```

```r
g1
```

## installing ggplot2 package

```r
install.packages('ggplot2')
```

```r
library(ggplot2)
```

## Graph to show category wise price for each item

```r
ggplot(data=d,aes(x=item,y=price,fill=category))+geom_bar(stat='identity')+coord_flip()+ggtitle('categorywise analysis')
```

## Graph showing pricewise Analysis

```r
price=ggplot(d,aes(x=item,y=price,fill=price))+geom_col()+coord_flip()+theme(legend.position='None',axis.text.y=element_text(size=6))+ggtitle('pricewise analysis')
```

```r
price
```
