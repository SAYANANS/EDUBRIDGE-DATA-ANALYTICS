```python
all_table=s.find_all('table') # to show table
print(all_table)
```

```python
model=s.find_all('div',{'class':'_4rR01T'})
```

```python
model
```

```python
product_name=[]
for i in model:
    p=i.text
    product_name.append(p)
```

```python
print(product_name)
```

```python
price=s.find_all('div',{'class':'_30jeq3 _1_WHN1'})
```

```python
prices=[]
for i in price:
    p=i.text
    prices.append(p)
```

```python
print(prices)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({'product Name':product_name,'Price':prices,'Rating':rates})
df.head(10)
```

## RATE VS PRICE

```python
plt.bar(df['Price'],df['Rating'],width=0.5)
plt.title('Rate Vs Price',fontsize=15,color='red')
plt.xlabel('price',fontsize=15)
plt.ylabel('Rate',fontsize=15)
plt.xticks(rotation=70)
plt.grid(True)
plt.show()
```

## COUNT THE PRODUCT

```python
sns.countplot(x='product Name',data=df,palette='coolwarm')
plt.xticks(rotation=90)
plt.show()
```
